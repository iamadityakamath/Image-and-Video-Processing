# Spatial Resolution & Downsampling

```
#Importing all the required Libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```
#Read the image
img2 = cv2.imread('/content/cameraman.tif',0)
```

```
#Determine the type of Image
type(img2)
```

```
#Get size of the image
[m,n]= img2.shape
print(m,n)
```

```
#Ask for sampling rate for down sampling from the user
f = int(input("Enter the down sampling rate: "))
```

```
#Downsample the image
img3=np.zeros((m//f,n//f),dtype=np.int)

for i in range(0,m,f):
  for j in range(0,n,f):
    try:
      img3[i//f][j//f] = img2[i][j]
    except IndexError:
      pass
```

```
img3.shape
```

```
#Writing the image
cv2.imwrite('sampled_img.png',img3)
```

```
plt.imshow(img2,cmap='gray')
```

```
plt.imshow(img3,cmap='gray')
```

```
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(2, 1, 1)
plt.imshow(img2,cmap='gray')
plt.xlabel("Original Image",color = 'b')

ax2 = fig.add_subplot(2, 1, 2)
plt.imshow(img3,cmap='gray')
plt.xlabel("Downscaled Image",color = 'b')

plt.show()
```

# Upscaling the Image

```
# to upsacle the image
img4=np.zeros((m , n),dtype=np.int)
for i in range(0, m-1, f):
  for j in range(0, n-1, f):
    img4[i,j]=img3[i//f][j//f]
```

```
plt.imshow(img4,cmap='gray')
```

```
#Nearest  neighbour Replication.

#Replication Rows
for i in range(1, m-1, f):
  for j in range(0, n-1):
    img4[i,j]=img4[i-1,j]
```

```
plt.imshow(img4,cmap='gray')
```

```
#Replication Columns
for i in range(0, m-1):
  for j in range(1, n-1, f):
    img4[i,j]=img4[i,j-1]
```

```
plt.imshow(img4,cmap='gray')
```

# Grey Level Resolution

```
# Finding Resolution of original Image
img5 = img2
[m,n]=img5.shape
print(m,n)
```

**Method 1** 

```
#Ask for the number of bit in which a pixel has to be represented in the new image
b= int(input("Enter the number of bits for the new image: "))

#Define the constants
maxold= 256
maxnew= 2**b
constant = maxnew/maxold

#creating new image matrix
img6= np.zeros((m,n), dtype= np.int) 

for i in range(m):
    for j in range(n):
        img6[i,j] = constant*img5[i,j]

#Plot Both the images
fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(2, 2, 1)
plt.imshow(img5,cmap='gray')
plt.xlabel("Original Image",color = 'b')

ax = fig.add_subplot(2, 2, 2)
plt.imshow(img6,cmap='gray')
plt.xlabel("Image with {} bit".format(b),color = 'r')
```

**Method 2**

```
# For a range of Bits
fig = plt.figure(figsize=(20, 20))

for b in range(0,8):
  img = np.zeros((m,n), dtype= np.int)
  for i in range(m):
    for j in range(n):
     img[i,j] = (2**b/256)*img5[i,j]

  ax = fig.add_subplot(5,5,b+1)
  plt.imshow(img,cmap='gray')
  plt.xlabel("Image with {} bits".format(b),color = 'b')
```

# Conclusion



* We change the grey level resolution of the image.
* We assign a no. of bits for the new image we want.
* We use the formula (2^b-1)/no. of bits of old image.
* The above is a constant we multiply to the intensity of each pixel.
* Thus we multiply the constant to each pixel in the old image to get the new image.
* We can also use for loop to assign no. of bits and compare iteratively.
* With increasing no. of bits, the image becomes brighter and quality of the image increases.
