# 1. Using Skimage Library.

```
from skimage import io
import matplotlib.pyplot as plt
img = io.imread('/content/lena_color_512.tif',plugin='pil')
#plt.savefig('img.png')
io.imshow(img)
```

# 2. Using CV2 Library.

```
#Read the image
import cv2
img1 = cv2.imread("/content/lena_color_512.tif")
cv2.imwrite("lena_color_256.tif",img1)
plt.imshow(img1)
```

```
#Determine the type of image
type(img1)
```

```
#To find the size of the image
img1.shape
```

```
#To find the Dimension of Array
img1.ndim
```

```
#To find the No. of Pixels/elements in the array
img1.size
```
